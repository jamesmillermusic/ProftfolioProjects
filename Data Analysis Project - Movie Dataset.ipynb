```python
# Dataset: https://www.kaggle.com/danielgrijalvas/movies

# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of the created plots


# Read in data

df = pd.read_csv(r'C:\Users\jamesmiller\Downloads\archive\movies.csv')
```

```python
# Looking at the data

df.head()
```

```python
# DATA CLEANING
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
# The output shows that all rows are complete
```

```python
# Check data types for columns

df.dtypes
```

```python
# Both budget and gross attriubtes do not need to be float, so we can change this to integer.
# Changing data types of 'budget' and 'gross' columns.

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# There are two years that appear in the dataset; 'released' and 'year'.
# These two should match, but there are some entries where this is not true, i.e. Row 7 - Platoon, 31 - River's Edge, 52 - Henry: Portrait of a Serial Killer
# To fix this issue: 1: create a new column, 2. prefer 'released' over 'year'


# Create Correct Year column

df['yearcorrect'] = df['released'].astype(str).str[:4]
```

```python
# Order by gross revenue

df.sort_values(by=['gross'], inplace=False, ascending=False).head()
```

```python
# View all entries, and not a condensed view of the data

pd.set_option('display.max_rows', None)
```

```python
# Remove duplicates

df['company'].drop_duplicates().sort_values(ascending=False)

# The output shows all the unique values within 'company'.
# If the above is executed without .drop_duplicates(), then there would be numerous entries for, example,'Zentropa Entertainments'
# Similarly, to drop all duplicates in the dataset, then the following should be executed: df.drop_duplicates()
```

```python
# CORRELATIONS
# Which attribute is more correlated to gross (revenue)

# Scatter plot: budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings (million, USD)')

plt.ylabel('Budget for Film (million, USD)')

plt.show()
```

```python
# Regression plot: Budget vs Gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color": "blue"})
```

```python
# Numerical version of the correlation

df.corr(method='pearson')
# methods: pearson, kendall, spearman
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Converting text columns into a numerical value

df_numeral = df

for col_name in df_numeral.columns:
    if(df_numeral[col_name].dtype == 'object'):
        df_numeral[col_name] = df_numeral[col_name].astype('category')
        df_numeral[col_name] = df_numeral[col_name].cat.codes

df_numeral.head()
```

```python
# Given there is now a numerical representation for all attributes, we can now provide a correlation matrix for all included data aspects of a movie

correlation_matrix = df_numeral.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()


# For example, the data show:
# Runtine has a 0.22 correlation with gross, genre has -0.35 correlation with budget.
```

```python
# To view output in, maybe, an easier manner

df_numeral.corr()
```

```python
# Organising the abouve correation output to see those with the higher correlation quicker.

# Unstacking will allow for each attribute to be aligned with each other (alphabetical) attribute and their correlation value.

correlation_mat = df_numeral.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Outlining highly positively correlated pairs

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
